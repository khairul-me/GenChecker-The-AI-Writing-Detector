```python
# GenChecker: AI-Generated Text Detection System
# For Kaggle competition: "LLM - Detect AI Generated Text"

# Install necessary packages
!pip install transformers torch pandas scikit-learn nltk spacy matplotlib seaborn textstat==0.7.3
!pip install accelerate bitsandbytes
!python -m spacy download en_core_web_sm
!pip install datasets evaluate plotly
import os
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import spacy
import textstat
import pickle
import time
import plotly.express as px
import plotly.graph_objects as go
from tqdm.auto import tqdm
from google.colab import files
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, confusion_matrix, roc_curve, precision_recall_curve
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
import torch
from datasets import Dataset
from IPython.display import HTML, display
import ipywidgets as widgets
from IPython.display import clear_output
import warnings
warnings.filterwarnings('ignore')

print("🔍 GenChecker: AI Text Detection System 🔍")
print("------------------------------------------")

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Create a data directory
!mkdir -p data

# Load spaCy model
print("Loading spaCy model...")
nlp = spacy.load("en_core_web_sm")
print("✅ spaCy model loaded successfully")

# Set random seed for reproducibility
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)

# Function to upload files
def upload_files():
    """Upload files through Colab interface"""
    print("Please upload the following Kaggle competition files:")
    print("- train_essays.csv")
    print("- test_essays.csv")
    print("- train_prompts.csv")
    print("- sample_submission.csv")

    uploaded = files.upload()

    for filename in uploaded.keys():
        with open(f'data/{filename}', 'wb') as f:
            f.write(uploaded[filename])
        print(f'✅ Saved {filename} to data directory')

# Function to explore dataset
def explore_dataset(file_path, n_rows=5):
    """Basic exploration of a CSV file"""
    try:
        df = pd.read_csv(file_path)
        print(f"\n📊 Exploring: {file_path}")
        print(f"Shape: {df.shape}")
        print(f"Columns: {df.columns.tolist()}")
        print(f"\nSample data:")
        print(df.head(n_rows))

        # Print additional info
        print("\nData types:")
        print(df.dtypes)

        if 'generated' in df.columns:
            print("\nClass distribution:")
            print(df['generated'].value_counts())
            print(f"Class balance: {df['generated'].value_counts(normalize=True)}")

        return df
    except Exception as e:
        print(f"❌ Error loading {file_path}: {e}")
        return None

# Function to load prompts data
def load_prompts_data(file_path):
    """Load prompt information from CSV file"""
    try:
        prompts_df = pd.read_csv(file_path)
        print(f"✅ Loaded {len(prompts_df)} prompts from {file_path}")

        # Display prompts info
        print("\nPrompt details:")
        for idx, row in prompts_df.iterrows():
            print(f"Prompt {row['prompt_id']}: {row['prompt_name']}")

        return prompts_df
    except Exception as e:
        print(f"❌ Error loading prompts data: {e}")
        return None

# Function to merge essays with their prompts for additional features
def enrich_essays_with_prompts(essays_df, prompts_df):
    """Add prompt information to essays dataframe"""
    if prompts_df is not None:
        print("Enriching essays with prompt information...")
        enriched_df = essays_df.merge(
            prompts_df,
            on='prompt_id',
            how='left'
        )
        print(f"✅ Essays enriched with {len(prompts_df)} prompts")
        return enriched_df
    return essays_df

# Function to extract linguistic features from text
def extract_features(text):
    """Extract linguistic features from text"""
    features = {}

    # Basic length features
    features['char_count'] = len(text)
    features['word_count'] = len(text.split())

    # Use spaCy for sentence tokenization
    try:
        doc = nlp(text[:50000])  # Limit text length to avoid memory issues
        sentences = list(doc.sents)
        features['sentence_count'] = len(sentences)

        if features['sentence_count'] > 0:
            features['avg_sentence_length'] = features['word_count'] / features['sentence_count']
        else:
            features['avg_sentence_length'] = 0

        # Word length features
        words = [token.text for token in doc if not token.is_punct]
        if words:
            features['avg_word_length'] = sum(len(word) for word in words) / len(words)
        else:
            features['avg_word_length'] = 0

        # Vocabulary diversity
        unique_words = set(words)
        features['unique_words'] = len(unique_words)
        if words:
            features['lexical_diversity'] = len(unique_words) / len(words)
        else:
            features['lexical_diversity'] = 0

        # Part of speech features
        pos_counts = {}
        for token in doc:
            pos = token.pos_
            pos_counts[pos] = pos_counts.get(pos, 0) + 1

        for pos, count in pos_counts.items():
            features[f'pos_{pos}'] = count / len(words) if words else 0

        # Entity features
        entity_counts = {}
        for ent in doc.ents:
            entity_counts[ent.label_] = entity_counts.get(ent.label_, 0) + 1

        for ent_type, count in entity_counts.items():
            features[f'ent_{ent_type}'] = count / len(sentences) if sentences else 0

    except Exception as e:
        # print(f"Error processing text: {e}")
        features['sentence_count'] = 0
        features['avg_sentence_length'] = 0
        features['avg_word_length'] = 0
        features['unique_words'] = 0
        features['lexical_diversity'] = 0

    # Readability scores
    try:
        features['flesch_reading_ease'] = textstat.flesch_reading_ease(text)
        features['flesch_kincaid_grade'] = textstat.flesch_kincaid_grade(text)
    except Exception as e:
        # print(f"Error with readability scores: {e}")
        features['flesch_reading_ease'] = 0
        features['flesch_kincaid_grade'] = 0

    return features

# Function to process essays with features
def process_essays_with_features(essays_df, batch_size=100):
    """Extract linguistic features from essays in smaller batches"""
    start_time = time.time()
    all_features = []

    print(f"Extracting features from {len(essays_df)} essays in batches of {batch_size}...")

    # Process in batches
    for i in range(0, len(essays_df), batch_size):
        batch_end = min(i + batch_size, len(essays_df))
        print(f"Processing batch {i//batch_size + 1}/{(len(essays_df) + batch_size - 1)//batch_size}: essays {i} to {batch_end-1}")

        batch_features = []
        # Process each essay in the current batch
        for idx in tqdm(range(i, batch_end)):
            row = essays_df.iloc[idx]
            text_features = extract_features(row['text'])
            batch_features.append(text_features)

        # Append batch results to the full list
        all_features.extend(batch_features)

        # Clear memory after each batch
        import gc
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    features_df = pd.DataFrame(all_features)
    result_df = pd.concat([essays_df.reset_index(drop=True), features_df], axis=1)

    elapsed_time = time.time() - start_time
    print(f"✅ Feature extraction complete! Time taken: {elapsed_time:.2f} seconds")

    return result_df

# Function to visualize features
def visualize_features(train_with_features, top_n=10):
    """Create visualizations for key features"""
    print("\n📊 Feature Visualizations 📊")

    # Select only numeric columns
    numeric_cols = train_with_features.select_dtypes(include=[np.number])

    # 1. Distribution of AI vs Human essays
    plt.figure(figsize=(10, 6))
    sns.countplot(x='generated', data=train_with_features)
    plt.title('Distribution of Human (0) vs AI (1) Generated Essays')
    plt.xlabel('Generated by AI')
    plt.ylabel('Count')
    plt.xticks([0, 1], ['Human', 'AI'])
    plt.show()

    # 2. Distribution across prompts
    plt.figure(figsize=(12, 6))
    prompt_counts = train_with_features['prompt_id'].value_counts()
    sns.barplot(x=prompt_counts.index, y=prompt_counts.values)
    plt.title('Distribution of Essays Across Prompts')
    plt.xlabel('Prompt ID')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

    # 3. Essay length comparison
    plt.figure(figsize=(12, 6))
    sns.histplot(data=train_with_features, x='word_count', hue='generated', kde=True, bins=30)
    plt.title('Word Count Distribution: Human vs AI')
    plt.xlabel('Word Count')
    plt.legend(['Human', 'AI'])
    plt.show()

    # 4. Lexical diversity comparison
    plt.figure(figsize=(12, 6))
    sns.histplot(data=train_with_features, x='lexical_diversity', hue='generated', kde=True, bins=30)
    plt.title('Lexical Diversity Distribution: Human vs AI')
    plt.xlabel('Lexical Diversity (unique words / total words)')
    plt.legend(['Human', 'AI'])
    plt.show()

    # 5. Correlation matrix of top features
    # Calculate correlation with target
    feature_correlations = numeric_cols.corrwith(train_with_features['generated']).sort_values(ascending=False)
    top_features = feature_correlations.abs().nlargest(top_n).index

    plt.figure(figsize=(14, 10))
    correlation_matrix = train_with_features[top_features].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title(f'Correlation Matrix of Top {top_n} Features')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # 6. Feature importance for distinguishing human vs AI
    plt.figure(figsize=(12, 8))
    top_discriminating = feature_correlations.abs().nlargest(15)
    sns.barplot(x=top_discriminating.values, y=top_discriminating.index)
    plt.title('Top 15 Features for Distinguishing Human vs AI Text')
    plt.xlabel('Correlation with AI-Generated Label')
    plt.tight_layout()
    plt.show()

    # 7. Interactive scatter plot with plotly
    fig = px.scatter(
        train_with_features,
        x='lexical_diversity',
        y='avg_sentence_length',
        color='generated',
        color_discrete_map={0: 'blue', 1: 'red'},
        labels={'generated': 'AI Generated'},
        title='Lexical Diversity vs Average Sentence Length',
        hover_data=['word_count', 'flesch_reading_ease']
    )
    fig.update_layout(legend_title_text='Generated by AI')
    fig.show()

    return feature_correlations

# Function to train a TF-IDF + LogReg model
def train_tfidf_model(train_essays):
    """Train and evaluate TF-IDF + Logistic Regression model"""
    print("\n🔍 Training TF-IDF + Logistic Regression Model")
    start_time = time.time()

    # Split data
    X_train, X_val, y_train, y_val = train_test_split(
        train_essays['text'],
        train_essays['generated'],
        test_size=0.2,
        random_state=RANDOM_SEED,
        stratify=train_essays['generated']
    )

    print(f"Training on {len(X_train)} samples, validating on {len(X_val)} samples")

    # Create a pipeline with TF-IDF and Logistic Regression
    pipeline = Pipeline([
        ('tfidf', TfidfVectorizer(max_features=10000, ngram_range=(1, 2))),
        ('clf', LogisticRegression(max_iter=1000, random_state=RANDOM_SEED, class_weight='balanced'))
    ])

    # Train the model
    pipeline.fit(X_train, y_train)

    # Evaluate
    y_pred = pipeline.predict(X_val)
    y_prob = pipeline.predict_proba(X_val)[:, 1]

    # Calculate metrics
    accuracy = accuracy_score(y_val, y_pred)
    roc_auc = roc_auc_score(y_val, y_prob)

    print(f"✅ TF-IDF + LogReg Model Performance:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_val, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_val, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix: TF-IDF + LogReg Model')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks([0.5, 1.5], ['Human', 'AI'])
    plt.yticks([0.5, 1.5], ['Human', 'AI'])
    plt.show()

    # ROC Curve
    plt.figure(figsize=(8, 6))
    fpr, tpr, _ = roc_curve(y_val, y_prob)
    plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve: TF-IDF + LogReg Model')
    plt.legend(loc='lower right')
    plt.grid(True, alpha=0.3)
    plt.show()

    elapsed_time = time.time() - start_time
    print(f"Training completed in {elapsed_time:.2f} seconds")

    return pipeline, X_val, y_val

# Function to train a Random Forest model on extracted features
def train_rf_model(train_with_features):
    """Train and evaluate Random Forest model on linguistic features"""
    print("\n🌲 Training Random Forest Model on Linguistic Features")
    start_time = time.time()

    # Get feature columns
    feature_cols = [col for col in train_with_features.columns if col not in ['id', 'prompt_id', 'text', 'generated', 'prompt_name', 'instructions', 'source_text']]

    # Split data
    X_train, X_val, y_train, y_val = train_test_split(
        train_with_features[feature_cols],
        train_with_features['generated'],
        test_size=0.2,
        random_state=RANDOM_SEED,
        stratify=train_with_features['generated']
    )

    print(f"Training on {len(X_train)} samples with {len(feature_cols)} features")

    # Train a Random Forest model
    rf_model = RandomForestClassifier(
        n_estimators=100,
        max_depth=20,
        min_samples_split=10,
        random_state=RANDOM_SEED,
        n_jobs=-1,
        class_weight='balanced'
    )
    rf_model.fit(X_train, y_train)

    # Evaluate
    y_pred = rf_model.predict(X_val)
    y_prob = rf_model.predict_proba(X_val)[:, 1]

    # Calculate metrics
    accuracy = accuracy_score(y_val, y_pred)
    roc_auc = roc_auc_score(y_val, y_prob)

    print(f"✅ Random Forest Model Performance:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_val, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_val, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
    plt.title('Confusion Matrix: Random Forest Model')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks([0.5, 1.5], ['Human', 'AI'])
    plt.yticks([0.5, 1.5], ['Human', 'AI'])
    plt.show()

    # ROC Curve
    plt.figure(figsize=(8, 6))
    fpr, tpr, _ = roc_curve(y_val, y_prob)
    plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve: Random Forest Model')
    plt.legend(loc='lower right')
    plt.grid(True, alpha=0.3)
    plt.show()

    # Feature importance analysis
    importances = rf_model.feature_importances_
    indices = np.argsort(importances)[::-1]
    feature_names = X_train.columns

    # Print top 20 features
    print("\nTop 20 most important features:")
    for i in range(min(20, len(feature_names))):
        print(f"{feature_names[indices[i]]}: {importances[indices[i]]:.4f}")

    # Plot feature importances
    plt.figure(figsize=(12, 8))
    plt.title("Random Forest Feature Importances")
    plt.bar(range(min(20, len(feature_names))), importances[indices[:20]], align='center')
    plt.xticks(range(min(20, len(feature_names))), [feature_names[i] for i in indices[:20]], rotation=90)
    plt.tight_layout()
    plt.show()

    elapsed_time = time.time() - start_time
    print(f"Training completed in {elapsed_time:.2f} seconds")

    return rf_model, feature_names, X_val, y_val

# Function to train a Transformer model
def train_transformer_model(train_essays):
    """Train and evaluate DeBERTa transformer model"""
    print("\n🤖 Training DeBERTa Transformer Model")
    start_time = time.time()

    # Prepare the dataset
    train_data = Dataset.from_pandas(pd.DataFrame({
        'text': train_essays['text'],
        'labels': train_essays['generated']
    }))
    train_data = train_data.train_test_split(test_size=0.2, seed=RANDOM_SEED)

    print(f"Training on {len(train_data['train'])} samples, validating on {len(train_data['test'])} samples")

    # Create the tokenizer
    tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-v3-small")

    def tokenize_function(examples):
        return tokenizer(examples['text'], padding='max_length', truncation=True, max_length=512)

    # Tokenize the dataset
    tokenized_train = train_data['train'].map(tokenize_function, batched=True)
    tokenized_val = train_data['test'].map(tokenize_function, batched=True)

    # Load model
    model = AutoModelForSequenceClassification.from_pretrained(
        "microsoft/deberta-v3-small",
        num_labels=2
    )

    # Set up training arguments
    training_args = TrainingArguments(
        output_dir='./results',
        learning_rate=2e-5,
        per_device_train_batch_size=2,  # Reduce from 4 to 2
        per_device_eval_batch_size=2,   # Reduce from 4 to 2
        num_train_epochs=3,
        weight_decay=0.01,
        report_to="none",
    )

    # Create trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_train,
        eval_dataset=tokenized_val,
        tokenizer=tokenizer,
    )

    # Train the model
    trainer.train()

    # Evaluate the model
    eval_results = trainer.evaluate()
    print(f"Transformer evaluation results: {eval_results}")

    # Save the model
    trainer.save_model("./transformer_model")

    # Get predictions on validation set
    predictions = trainer.predict(tokenized_val)
    y_prob = torch.softmax(torch.tensor(predictions.predictions), dim=1)[:, 1].numpy()
    y_pred = np.argmax(predictions.predictions, axis=1)
    y_true = train_data['test']['labels']

    # Calculate metrics
    accuracy = accuracy_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_prob)

    print(f"✅ Transformer Model Performance:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Purples')
    plt.title('Confusion Matrix: Transformer Model')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks([0.5, 1.5], ['Human', 'AI'])
    plt.yticks([0.5, 1.5], ['Human', 'AI'])
    plt.show()

    # ROC Curve
    plt.figure(figsize=(8, 6))
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve: Transformer Model')
    plt.legend(loc='lower right')
    plt.grid(True, alpha=0.3)
    plt.show()

    elapsed_time = time.time() - start_time
    print(f"Training completed in {elapsed_time:.2f} seconds")

    return model, tokenizer, tokenized_val, train_data['test']

# Function to get ensemble predictions
def get_ensemble_predictions(text_data, feature_data, tfidf_pipeline, rf_model, feature_names, transformer_model, tokenizer):
    """Combine predictions from all models"""
    print("Generating ensemble predictions...")

    # Get TF-IDF + LogReg predictions
    tfidf_preds = tfidf_pipeline.predict_proba(text_data)[:, 1]

    # Get feature-based predictions
    # Ensure feature data has exactly the same columns in the same order
    # First, make sure all required columns exist
    for col in feature_names:
        if col not in feature_data.columns:
            feature_data[col] = 0  # Add missing columns with default values

    # Select only the columns used during training and in the right order
    rf_preds = rf_model.predict_proba(feature_data[feature_names])[:, 1]

    # Get transformer predictions
    inputs = tokenizer(list(text_data), padding='max_length', truncation=True, max_length=512, return_tensors="pt")
    with torch.no_grad():
        outputs = transformer_model(**{k: v.to(transformer_model.device) for k, v in inputs.items()})
    transformer_preds = torch.softmax(outputs.logits, dim=1)[:, 1].cpu().numpy()

    # Combine predictions with weighted average
    # You could experiment with different weights
    ensemble_preds = (0.3 * tfidf_preds + 0.3 * rf_preds + 0.4 * transformer_preds)

    print("✅ Ensemble predictions complete!")

    return ensemble_preds, tfidf_preds, rf_preds, transformer_preds

# Function to visualize ensemble performance
def visualize_ensemble_performance(y_true, tfidf_preds, rf_preds, transformer_preds, ensemble_preds):
    """Visualize and compare all model predictions"""
    print("\n📊 Ensemble Model Performance Visualization")

    # Convert probabilities to binary predictions
    tfidf_binary = (tfidf_preds > 0.5).astype(int)
    rf_binary = (rf_preds > 0.5).astype(int)
    transformer_binary = (transformer_preds > 0.5).astype(int)
    ensemble_binary = (ensemble_preds > 0.5).astype(int)

    # Calculate metrics
    metrics = {
        'Model': ['TF-IDF + LogReg', 'Random Forest', 'Transformer', 'Ensemble'],
        'Accuracy': [
            accuracy_score(y_true, tfidf_binary),
            accuracy_score(y_true, rf_binary),
            accuracy_score(y_true, transformer_binary),
            accuracy_score(y_true, ensemble_binary)
        ],
        'ROC AUC': [
            roc_auc_score(y_true, tfidf_preds),
            roc_auc_score(y_true, rf_preds),
            roc_auc_score(y_true, transformer_preds),
            roc_auc_score(y_true, ensemble_preds)
        ]
    }

    metrics_df = pd.DataFrame(metrics)
    print("Model Performance Comparison:")
    print(metrics_df)

    # Bar chart comparing metrics
    plt.figure(figsize=(12, 6))

    # Accuracy comparison
    plt.subplot(1, 2, 1)
    sns.barplot(x='Model', y='Accuracy', data=metrics_df)
    plt.title('Accuracy Comparison')
    plt.ylim(0.5, 1.0)
    plt.xticks(rotation=45)

    # AUC comparison
    plt.subplot(1, 2, 2)
    sns.barplot(x='Model', y='ROC AUC', data=metrics_df)
    plt.title('ROC AUC Comparison')
    plt.ylim(0.5, 1.0)
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

    # ROC curves comparison
    plt.figure(figsize=(10, 8))
    # TF-IDF ROC
    fpr_tfidf, tpr_tfidf, _ = roc_curve(y_true, tfidf_preds)
    plt.plot(fpr_tfidf, tpr_tfidf, label=f'TF-IDF + LogReg (AUC = {metrics["ROC AUC"][0]:.4f})')

    # RF ROC
    fpr_rf, tpr_rf, _ = roc_curve(y_true, rf_preds)
    plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {metrics["ROC AUC"][1]:.4f})')

    # Transformer ROC
    fpr_transformer, tpr_transformer, _ = roc_curve(y_true, transformer_preds)
    plt.plot(fpr_transformer, tpr_transformer, label=f'Transformer (AUC = {metrics["ROC AUC"][2]:.4f})')

    # Ensemble ROC
    fpr_ensemble, tpr_ensemble, _ = roc_curve(y_true, ensemble_preds)
    plt.plot(fpr_ensemble, tpr_ensemble, label=f'Ensemble (AUC = {metrics["ROC AUC"][3]:.4f})')

    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve Comparison')
    plt.legend(loc='lower right')
    plt.grid(True, alpha=0.3)
    plt.show()

    # Confusion Matrix for Ensemble
    cm_ensemble = confusion_matrix(y_true, ensemble_binary)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_ensemble, annot=True, fmt='d', cmap='Oranges')
    plt.title('Confusion Matrix: Ensemble Model')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks([0.5, 1.5], ['Human', 'AI'])
    plt.yticks([0.5, 1.5], ['Human', 'AI'])
    plt.show()

    # Precision-Recall curve
    plt.figure(figsize=(10, 8))

    # TF-IDF PR curve
    precision_tfidf, recall_tfidf, _ = precision_recall_curve(y_true, tfidf_preds)
    plt.plot(recall_tfidf, precision_tfidf, label='TF-IDF + LogReg')

    # RF PR curve
    precision_rf, recall_rf, _ = precision_recall_curve(y_true, rf_preds)
    plt.plot(recall_rf, precision_rf, label='Random Forest')

    # Transformer PR curve
    precision_transformer, recall_transformer, _ = precision_recall_curve(y_true, transformer_preds)
    plt.plot(recall_transformer, precision_transformer, label='Transformer')

    # Ensemble PR curve
    precision_ensemble, recall_ensemble, _ = precision_recall_curve(y_true, ensemble_preds)
    plt.plot(recall_ensemble, precision_ensemble, label='Ensemble')

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve Comparison')
    plt.legend(loc='best')
    plt.grid(True, alpha=0.3)
    plt.show()

    return metrics_df

# Function to analyze a single text
def analyze_text(text, rf_model, feature_names, tfidf_pipeline, transformer_model, tokenizer):
    """Extract features and predict whether text is AI-generated"""
    # Extract features
    features = extract_features(text)
    features_df = pd.DataFrame([features])

    # Make sure all feature columns exist
    for col in feature_names:
        if col not in features_df.columns:
            features_df[col] = 0

    # Get individual model predictions
    rf_pred = rf_model.predict_proba(features_df[feature_names])[0, 1]
    tfidf_pred = tfidf_pipeline.predict_proba([text])[0, 1]

    inputs = tokenizer([text], padding='max_length', truncation=True, max_length=512, return_tensors="pt")
    with torch.no_grad():
        outputs = transformer_model(**{k: v.to(transformer_model.device) for k, v in inputs.items()})
    transformer_pred = torch.softmax(outputs.logits, dim=1)[0, 1].cpu().numpy()

    # Get ensemble prediction
    ensemble_pred = (0.3 * rf_pred + 0.3 * tfidf_pred + 0.4 * transformer_pred)

    # Return prediction and key features
    return {
        'ensemble_pred': ensemble_pred,
        'rf_pred': rf_pred,
        'tfidf_pred': tfidf_pred,
        'transformer_pred': transformer_pred,
        'features': {
            'word_count': features.get('word_count', 0),
            'char_count': features.get('char_count', 0),
            'unique_words': features.get('unique_words', 0),
            'lexical_diversity': features.get('lexical_diversity', 0),
            'sentence_count': features.get('sentence_count', 0),
            'avg_sentence_length': features.get('avg_sentence_length', 0),
            'avg_word_length': features.get('avg_word_length', 0),
            'flesch_reading_ease': features.get('flesch_reading_ease', 0),
            'flesch_kincaid_grade': features.get('flesch_kincaid_grade', 0)
        }
    }

# Function to create interactive detector interface
def create_detector_interface(rf_model, feature_names, tfidf_pipeline, transformer_model, tokenizer):
    """Create an interactive UI for analyzing text"""
    print("\n✨ Launching AI Text Detector Interface")

    # Create text area for input
    text_area = widgets.Textarea(
        value='',
        placeholder='Enter your text here to check if it was written by AI...',
        description='',
        disabled=False,
        layout=widgets.Layout(width='100%', height='200px')
    )

    # Create button for analysis
    button = widgets.Button(
        description='Analyze Text',
        button_style='primary',
        tooltip='Click to analyze the text',
        icon='check'
    )

    # Create output area
    output = widgets.Output()

    # Define button click event
    def on_button_clicked(b):
        with output:
            clear_output()
            if not text_area.value.strip():
                print("Please enter some text to analyze.")
                return

            # Analyze the text
            result = analyze_text(
                text_area.value,
                rf_model,
                feature_names,
                tfidf_pipeline,
                transformer_model,
                tokenizer
            )
            ai_score = result['ensemble_pred'] * 100

            # Bar color - red for AI, green for human
            bar_color = "#d63031" if ai_score > 50 else "#00b894"

            # Display results with improved formatting
            html_output = f"""
            <div style="background-color: #f8f9fa; padding: 20px; border-radius: 10px; margin-bottom: 20px; font-family: Arial, sans-serif;">
                <h2 style="text-align: center; color: #2d3436; font-size: 24px; margin-bottom: 20px;">AI Text Detector Results</h2>
                <div style="text-align: center; margin: 20px 0;">
                    <span style="font-size: 28px; font-weight: bold; color: #2d3436;">{ai_score:.1f}% AI content</span>
                </div>
                <div style="background-color: #dfe6e9; height: 30px; width: 100%; border-radius: 5px; margin-bottom: 10px; position: relative;">
                    <div style="background-color: {bar_color}; height: 30px; width: {ai_score}%; border-radius: 5px; position: absolute; top: 0; left: 0;"></div>
                </div>
                <div style="display: flex; justify-content: space-between; color: #2d3436; font-weight: bold; margin-bottom: 20px;">
                    <span>0% AI</span>
                    <span>100% AI</span>
                </div>
                <h3 style="margin-top: 20px; color: #2d3436; font-size: 20px;">Model Breakdown:</h3>
                <ul style="list-style-type: none; padding-left: 10px; font-size: 16px; color: #2d3436;">
                    <li style="margin-bottom: 8px;"><strong>Random Forest:</strong> <span style="font-weight: bold;">{result['rf_pred']*100:.1f}%</span></li>
                    <li style="margin-bottom: 8px;"><strong>TF-IDF + LogReg:</strong> <span style="font-weight: bold;">{result['tfidf_pred']*100:.1f}%</span></li>
                    <li style="margin-bottom: 8px;"><strong>Transformer:</strong> <span style="font-weight: bold;">{result['transformer_pred']*100:.1f}%</span></li>
                </ul>
                <h3 style="margin-top: 20px; color: #2d3436; font-size: 20px;">Key Linguistic Features:</h3>
                <ul style="list-style-type: none; padding-left: 10px; font-size: 16px; color: #2d3436;">
                    <li style="margin-bottom: 8px;"><strong>Word count:</strong> <span style="font-weight: bold;">{result['features']['word_count']:.0f}</span></li>
                    <li style="margin-bottom: 8px;"><strong>Lexical diversity:</strong> <span style="font-weight: bold;">{result['features']['lexical_diversity']:.4f}</span></li>
                    <li style="margin-bottom: 8px;"><strong>Average sentence length:</strong> <span style="font-weight: bold;">{result['features']['avg_sentence_length']:.2f}</span></li>
                    <li style="margin-bottom: 8px;"><strong>Average word length:</strong> <span style="font-weight: bold;">{result['features']['avg_word_length']:.2f}</span></li>
                    <li style="margin-bottom: 8px;"><strong>Flesch reading ease:</strong> <span style="font-weight: bold;">{result['features']['flesch_reading_ease']:.2f}</span></li>
                </ul>
            </div>
            """
            display(HTML(html_output))

    button.on_click(on_button_clicked)

    # Display the interface with improved styling
    display(widgets.HTML(value="<h1 style='text-align: center; font-family: Arial, sans-serif; color: #2d3436;'>GenChecker: AI Content Detector</h1>"))
    display(widgets.HTML(value="<p style='text-align: center; font-family: Arial, sans-serif; color: #2d3436; font-size: 16px;'>Analyze text to determine if it was written by AI</p>"))
    display(text_area)
    display(button)
    display(output)

# Function to test on example texts
def test_on_examples(rf_model, feature_names, tfidf_pipeline, transformer_model, tokenizer):
    """Test the ensemble model on sample texts"""
    print("\n🧪 Testing Ensemble Model on Example Texts")

    # Example texts to test
    human_text = """
    The sunset painted the sky with hues of orange and purple, a masterpiece that no artist could replicate. As I stood on the cliff overlooking the ocean, I felt a sense of peace wash over me. The gentle breeze carried the scent of salt and seaweed, reminding me of childhood summers spent building sandcastles and chasing waves.
    """

    ai_text = """
    The sunset exhibited a chromatic display across the atmospheric canvas, presenting a visual phenomenon that artistic reproduction cannot fully capture. Upon positioning myself at the precipice adjacent to the oceanic expanse, a sensation of tranquility permeated my consciousness. The atmospheric current transported olfactory particles of sodium chloride and marine vegetation, evoking mnemonic connections to juvenile estival periods dedicated to the construction of sand-based architectural structures and the pursuit of hydrodynamic oscillations.
    """

    print("\n===== HUMAN-WRITTEN TEXT ANALYSIS =====")
    human_result = analyze_text(human_text, rf_model, feature_names, tfidf_pipeline, transformer_model, tokenizer)
    print(f"Ensemble prediction: {human_result['ensemble_pred']:.4f}")
    print(f"RF model: {human_result['rf_pred']:.4f}")
    print(f"TF-IDF model: {human_result['tfidf_pred']:.4f}")
    print(f"Transformer model: {human_result['transformer_pred']:.4f}")

    print("\nKey features:")
    for feature, value in human_result['features'].items():
        print(f"{feature}: {value:.4f}")

    print("\n===== AI-GENERATED TEXT ANALYSIS =====")
    ai_result = analyze_text(ai_text, rf_model, feature_names, tfidf_pipeline, transformer_model, tokenizer)
    print(f"Ensemble prediction: {ai_result['ensemble_pred']:.4f}")
    print(f"RF model: {ai_result['rf_pred']:.4f}")
    print(f"TF-IDF model: {ai_result['tfidf_pred']:.4f}")
    print(f"Transformer model: {ai_result['transformer_pred']:.4f}")

    print("\nKey features:")
    for feature, value in ai_result['features'].items():
        print(f"{feature}: {value:.4f}")

    # Compare features
    print("\n===== FEATURE COMPARISON =====")
    comparison = {
        'Feature': list(human_result['features'].keys()),
        'Human': [human_result['features'][k] for k in human_result['features']],
        'AI': [ai_result['features'][k] for k in ai_result['features']]
    }
    comparison_df = pd.DataFrame(comparison)
    print(comparison_df)

    # Visualize feature differences
    plt.figure(figsize=(12, 8))
    for i, feature in enumerate(['lexical_diversity', 'avg_sentence_length', 'avg_word_length', 'flesch_reading_ease']):
        plt.subplot(2, 2, i+1)
        values = [human_result['features'][feature], ai_result['features'][feature]]
        plt.bar(['Human', 'AI'], values, color=['#00b894', '#d63031'])
        plt.title(f'{feature.replace("_", " ").title()}')
        plt.ylabel('Value')

    plt.tight_layout()
    plt.show()

    return human_result, ai_result

# Function to process test data and generate submission
def generate_kaggle_submission(
    test_essays,
    rf_model,
    feature_names,
    tfidf_pipeline,
    transformer_model,
    tokenizer,
    sample_submission
):
    """Process test data and create submission file for Kaggle"""
    print("\n🏆 Generating Kaggle Submission")
    print(f"Processing {len(test_essays)} test essays...")
    start_time = time.time()

    # Extract features from test essays
    test_with_features = process_essays_with_features(test_essays)

    # Get ensemble predictions
    text_data = test_essays['text']
    ensemble_preds, tfidf_preds, rf_preds, transformer_preds = get_ensemble_predictions(
        text_data,
        test_with_features,
        tfidf_pipeline,
        rf_model,
        feature_names,
        transformer_model,
        tokenizer
    )

    # Convert probabilities to binary predictions (0 or 1)
    binary_preds = (ensemble_preds > 0.5).astype(int)

    # Create submission file
    submission = pd.DataFrame({
        'id': test_essays['id'],
        'generated': binary_preds  # Use binary predictions for final submission
    })

    # Ensure submission format matches sample_submission
    if sample_submission is not None:
        if set(submission['id']) != set(sample_submission['id']):
            print("⚠️ Warning: Submission IDs don't match sample_submission IDs")

        # Reindex to match sample_submission order
        submission = submission.set_index('id')
        submission = submission.reindex(index=sample_submission['id'])
        submission = submission.reset_index()

    # Save submission file
    submission.to_csv('submission.csv', index=False)

    elapsed_time = time.time() - start_time
    print(f"✅ Submission file created: submission.csv")
    print(f"Time taken: {elapsed_time:.2f} seconds")

    # Show submission preview
    print("\nSubmission preview:")
    print(submission.head())

    # Show prediction distribution
    plt.figure(figsize=(8, 6))
    sns.countplot(x='generated', data=submission)
    plt.title('Distribution of Predictions')
    plt.xlabel('Generated by AI')
    plt.xticks([0, 1], ['Human', 'AI'])
    plt.ylabel('Count')
    plt.show()

    # Show probability distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(ensemble_preds, bins=30, kde=True)
    plt.axvline(x=0.5, color='red', linestyle='--', label='Decision Threshold')
    plt.title('Distribution of Ensemble Prediction Probabilities')
    plt.xlabel('Probability of being AI-generated')
    plt.ylabel('Count')
    plt.legend()
    plt.show()

    return submission

# Main function for Kaggle submission
def main_kaggle():
    """Main function to train models and generate Kaggle submission"""
    print("\n🚀 Starting GenChecker AI Text Detection System")
    print("================================================")

    # Upload files if needed
    required_files = ['train_essays.csv', 'test_essays.csv', 'train_prompts.csv', 'sample_submission.csv']
    missing_files = [f for f in required_files if not os.path.exists(f'data/{f}')]

    if missing_files:
        print(f"Missing files: {missing_files}")
        upload_files()

    # Load data
    train_essays = explore_dataset('data/train_essays.csv')
    test_essays = explore_dataset('data/test_essays.csv')
    train_prompts = load_prompts_data('data/train_prompts.csv')
    sample_submission = pd.read_csv('data/sample_submission.csv')

    print(f"Loaded {len(train_essays)} training essays and {len(test_essays)} test essays")
    print(f"Submission format requires {len(sample_submission)} predictions")

    # Optional: Enrich essays with prompt information
    # This can be useful for feature engineering based on prompts
    # train_essays = enrich_essays_with_prompts(train_essays, train_prompts)
    # test_essays = enrich_essays_with_prompts(test_essays, train_prompts)

    # Process essays with features
    print("\n🔎 Extracting Linguistic Features")
    train_with_features = process_essays_with_features(train_essays)

    # Visualize features
    feature_correlations = visualize_features(train_with_features)

    # Train models
    print("\n🏗️ Building and Training Models")

    # 1. TF-IDF + LogReg Model
    tfidf_pipeline, tfidf_val_texts, tfidf_val_labels = train_tfidf_model(train_essays)

    # 2. Random Forest Model
    rf_model, feature_names, rf_val_feats, rf_val_labels = train_rf_model(train_with_features)

    # 3. Transformer Model
    transformer_model, tokenizer, transformer_val_data, transformer_val_labels = train_transformer_model(train_essays)

    # Save models (optional but recommended)
    print("\n💾 Saving Models")
    with open('rf_model.pkl', 'wb') as f:
        pickle.dump(rf_model, f)
    with open('feature_names.pkl', 'wb') as f:
        pickle.dump(feature_names, f)
    with open('tfidf_pipeline.pkl', 'wb') as f:
        pickle.dump(tfidf_pipeline, f)
    print("Models saved successfully!")

    # Test on validation set
    print("\n🧪 Evaluating Ensemble on Validation Set")
    # Get predictions on validation set for ensemble evaluation
    val_X = train_essays.iloc[tfidf_val_texts.index]['text']  # Use same validation indices as tfidf model
    val_features = train_with_features.iloc[tfidf_val_texts.index]
    val_y = tfidf_val_labels  # Labels for validation set

    # Get ensemble predictions on validation set
    val_ensemble_preds, val_tfidf_preds, val_rf_preds, val_transformer_preds = get_ensemble_predictions(
        val_X,
        val_features,
        tfidf_pipeline,
        rf_model,
        feature_names,
        transformer_model,
        tokenizer
    )

    # Visualize ensemble performance
    ensemble_metrics = visualize_ensemble_performance(
        val_y,
        val_tfidf_preds,
        val_rf_preds,
        val_transformer_preds,
        val_ensemble_preds
    )

    # Test on example texts
    human_example, ai_example = test_on_examples(rf_model, feature_names, tfidf_pipeline, transformer_model, tokenizer)

    # Generate Kaggle submission
    submission = generate_kaggle_submission(
        test_essays,
        rf_model,
        feature_names,
        tfidf_pipeline,
        transformer_model,
        tokenizer,
        sample_submission
    )

    # Create interactive interface
    print("\n🎮 Creating Interactive Interface")
    create_detector_interface(rf_model, feature_names, tfidf_pipeline, transformer_model, tokenizer)

    print("\n✅ GenChecker AI Text Detection System is ready!")
    print("Kaggle submission has been generated: submission.csv")

    return {
        'models': {
            'rf_model': rf_model,
            'tfidf_pipeline': tfidf_pipeline,
            'transformer_model': transformer_model,
            'tokenizer': tokenizer,
            'feature_names': feature_names
        },
        'data': {
            'train_essays': train_essays,
            'test_essays': test_essays,
            'train_with_features': train_with_features
        },
        'results': {
            'ensemble_metrics': ensemble_metrics,
            'submission': submission
        }
    }

# Run the Kaggle-specific main function
if __name__ == "__main__":
    main_kaggle()
```
